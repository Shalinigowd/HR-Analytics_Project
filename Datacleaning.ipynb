```python
#Import Required Libraries

import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("HR_analytacis.csv")
df.head()
```

```python
print(df.shape)
print(df.info())
print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
df['MonthlyIncome'] = df['MonthlyIncome'].fillna(df['MonthlyIncome'].mean())
```

```python
print(df.duplicated().sum())
```

```python
df.columns = df.columns.str.lower()

df.rename(columns={
    'monthlyincome':'salary',
    'jobrole':'job_role'
}, inplace=True)
print(df.columns)
```

```python
df.info()
```

```python
#handeling Outliears

Q1 = df['salary'].quantile(0.25)
Q3 = df['salary'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['salary'] < lower) | (df['salary'] > upper)]

print(outliers)
```

```python
df = df[(df['salary'] >= lower) & (df['salary'] <= upper)]
```

```python
df['gender'] = df['gender'].replace({
    'M':'Male',
    'F':'Female'
})
```

```python
df['department'] = df['department'].str.strip()
```

```python
df['department'] = df['department'].str.title()

df['job_role'] = df['job_role'].str.upper()
```

```python
df['overtime'] = df['overtime'].replace({
    'Yes':'Y',
    'No':'N'
})
```

```python
print(df['department'].unique())
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['attrition'] = le.fit_transform(df['attrition'])
```

```python
df = pd.get_dummies(df, columns=['department'])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df['salary'] = scaler.fit_transform(df[['salary']])
```

```python
from sklearn.preprocessing import MinMaxScaler

mm = MinMaxScaler()

df['age'] = mm.fit_transform(df[['age']])
```

```python
corr = df.corr(numeric_only=True)

print(corr['attrition'].sort_values(ascending=False))
```

```python
null_rows = df[df.isnull().any(axis=1)]

print(null_rows)
```

```python
print(df.corr(numeric_only=True))
```

```python
df.to_csv("Cleaned_HR_Analytics.csv", index=False)
```
